```python
# Avant migration (DataFrame)
print("Colonnes :", df.columns.tolist())
print("Types :", df.dtypes)
print("Doublons :", df.duplicated().sum())
print("Valeurs manquantes :\n", df.isnull().sum())

# Après migration (MongoDB)
sample = collection.find_one()
print("Clés MongoDB :", sample.keys())

# Exemple pour compter les valeurs manquantes dans MongoDB pour la colonne 'Name'
missing_names = collection.count_documents({"Name": {"$in": [None, ""]}})
print("Valeurs manquantes 'Name' dans MongoDB :", missing_names)

# Exemple pour compter les doublons sur 'Name' + 'Date of Admission'
pipeline = [
    {"$group": {
        "_id": {"Name": "$Name", "Date of Admission": "$Date of Admission"},
        "count": {"$sum": 1}
    }},
    {"$match": {"count": {"$gt": 1}}}
]
duplicates = list(collection.aggregate(pipeline))
print("Doublons MongoDB :", len(duplicates))
```
